```python
x=int(input ("enter the first number:"))
y=int(input("enter the second number:"))
op=input("enter your choice of operation:")
match op:
  case '+':
    a = module1.add(x,y)
    print(a)
  case '-':
    b=module1.sub(x,y)
    print(b)
  case '*':
    c=module1.mul(x,y)
    print(c)
  case '/':
    d=module1.div(x,y)
    print(d)
  case '%':
    e=module1.modulo(x,y)
    print(e)
  case '_':
    print("invalid")


```
